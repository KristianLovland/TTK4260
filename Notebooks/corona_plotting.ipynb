## Coronavirus spread visualization
More specifically, norwegian infection numbers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corona_dataset_timeseries_global_path = "../datasett/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
corona_data = pd.read_csv(corona_dataset_timeseries_global_path)
corona_data.set_index('Country/Region', inplace=True)
```

```python
infected_norwegians = corona_data.loc['Norway', :]
# Omit first three columns consisting of county (n/a) and location
infected_norwegians = infected_norwegians.iloc[3:len(infected_norwegians)].values
time = np.linspace(1, len(infected_norwegians), len(infected_norwegians))
infected_norwegians = np.vstack((time, infected_norwegians))
infected_norwegians = pd.DataFrame(data=infected_norwegians, index=["Time", "Infected norwegians"], dtype=float)
infected_norwegians =  infected_norwegians.transpose()
```

```python
%config InlineBackend.figure_format = 'svg'

sns.set_context("notebook")
sns.set(style="ticks", font="Latin Modern Math")

plt.close()
plt.figure(figsize=(12,6))
sns.lineplot(x="Time", y="Infected norwegians", data=infected_norwegians);
```

```python
sns.set_context("notebook")
sns.set(style="ticks", font="Times New Roman")

plt.figure(figsize=(16, 6))
sns.lineplot(x="Time", y="Norway", data=infected_norwegians)
```
